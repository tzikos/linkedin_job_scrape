```python
import pandas as pd
import PyPDF2
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Open the PDF file
pdf_file = open('./CV_PAPANTZIKOS_DIMITRIOS.pdf', 'rb')

# Create a PDF object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Initialize an empty string to store the text
text = ''

# Loop through each page and extract text
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages[page_num]
    text += page.extract_text()

# Close the PDF file
pdf_file.close()

# Print or manipulate the extracted text
print(text)
```

```python
jobs = pd.read_csv('./10204_LinkedIn_jobs.csv')
```

```python
jobs['id'] = [i for i in range(len(jobs))]
```

```python
text_contents = jobs['Job Descr. text']
filenames = jobs['id']
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(text_contents)
```

```python
def query1(query_text):
    query_vector = tfidf_vectorizer.transform([query_text])
    similarities = cosine_similarity(query_vector, tfidf_matrix)
    top_indices = np.argsort(similarities[0])[::-1]

    if similarities[0][top_indices[0]] == 0:
        print('No match.')
        top_filenames = []
    else:
        top_filenames = [filenames[idx] for idx in top_indices if similarities[0][idx] > 0]
    return top_filenames
```

```python
results = query1(text)
```

```python
result_df = jobs.iloc[results,:]
final = result_df[~result_df['Work Times'].str.contains('Senior') & ~result_df['Work Times'].str.contains('Director') & ~result_df['Work Times'].str.contains('Manager') & ~result_df['Work Times'].str.contains('Lead') & ~result_df['Work Times'].str.contains('Head')]
final
```
